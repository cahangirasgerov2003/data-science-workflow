```python
import pandas as pd
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt
```

# Element wise multiplication or HADAMARD product

```python
rng = np.random.default_rng(0)
a1 = rng.integers(1, 10, size=(5, 3))
a2 = rng.integers(1, 10, size=(5, 3))
a1, a2
```

```python
a1 * a2
```

# Dot product

```python
np.dot(a1, a2)
```

```python
a1.shape, a2.shape
```

```python
transposed_a2 = a2.T
```

```python
np.dot(a1, transposed_a2)
```
