```python
import pandas as pd
```

```python
car_datas = pd.read_csv("car-sales.csv")
```

```python
car_datas
```

```python
# Shuffling data 
shuffled_data = car_datas.sample(frac=1)
```

```python
shuffled_data
```

```python
# Does not show indexes created when there is a shuffle
shuffled_data.reset_index(drop=True)
```

```python
shuffled_data['Odometer (KM)'] = shuffled_data['Odometer (KM)'].apply(lambda x:x / 1.6).round()
```
