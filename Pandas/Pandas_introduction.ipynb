## 2 main data types 

### Series and dataframes

```python
import pandas as pd
```

```python
car_brands = pd.Series(["BMW", "Mercedes", "Opel", "Audi"])
```

```python
car_brands
```

#### Series ----- 1 dimensional 

```python
car_colours = pd.Series(["Red", "Black", "Gray"])
```

#### DataFrames ----- 2 dimensional

```python
car_datas = pd.DataFrame({"car_brands": car_brands, "car_colours": car_colours})
```

```python
car_datas
```
