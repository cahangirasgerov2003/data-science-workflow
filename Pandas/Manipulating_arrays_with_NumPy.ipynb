# Aggreagtion

```python
rng = np.random.default_rng(0)
a1 = rng.integers(1, 100, 100000)
```

```python
a1[:10]
```

```python
# Python sum() method
%timeit sum(a1)
# Np.sum() method
%timeit np.sum(a1)
```

```python
# Standart deviation 
np.std(a2)
```
