# Manipulating data with pandas 2

```python
car_datas = pd.read_csv("car-sales.csv")
car_datas
```

```python
car_datas["Price"] = car_datas['Price'].astype(str).str.replace(r"[^\d.]", "", regex = True).astype(float).round().astype(int)
```

```python
car_datas
```

```python
car_sales_missing_data
```

```python
car_sales_missing_data.dropna()
```

```python
# Column from Series
gasoline_consumption_per_km = pd.Series([10, 9, 5, 30, 1])
car_datas["Gasoline_consumption_per_km"] = gasoline_consumption_per_km
```

```python
car_datas
```

```python
car_datas['Gasoline_consumption_per_km'] = car_datas["Gasoline_consumption_per_km"].fillna(5)
```

```python
car_datas
```

```python
# Column from Python List
car_seats = []
for i in range(len(car_datas)):
    car_seats.append(5)
car_datas['Seats'] = car_seats
```

```python
car_datas
```

```python
# 
car_datas["Gasoline consumption so far (L)"] = car_datas['Odometer (KM)'] * car_datas['Gasoline_consumption_per_km'] / 100
```

```python
car_datas
```

```python
car_datas  = car_datas.drop("gasoline_consumption_per_km", axis=1)
```

```python
car_datas
```

```python
# Create column from a single value
car_datas["Wheels"] = 4
```
