```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Std() & var() methods
```

```python
rng = np.random.default_rng(0)
a1 = rng.integers(0, 10, size = 10)
a1
```

```python
rng = np.random.default_rng(0)
a2 = rng.integers(0, 1000, size = 10)
a2
```

```python
np.var(a1), np.var(a2)
```

```python
np.std(a1), np.std(a2)
```

```python
plt.hist(a1)
plt.show
```

```python
plt.hist(a2)
plt.show
```
