### Getting the data ready

```python
import warnings
warnings.filterwarnings(action = "error")
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import sklearn
```

```python
# Check python version
sklearn.show_versions()
```

```python
car_sales_missing_df = pd.read_csv("./data/car-sales-extended-missing-data.csv")
car_sales_missing_df.head()
```

```python
len(car_sales_missing_df)
```

```python
# How many missing values are there ?
car_sales_missing_df.isna().sum()
```

```python
# Option 1: Handling missing values with pandas
car_sales_missing_df["Make"].fillna("Missing", inplace=True)

car_sales_missing_df["Colour"].fillna("Missing", inplace=True)

car_sales_missing_df["Doors"].fillna(4, inplace=True)

car_sales_missing_df['Odometer (KM)'].fillna(car_sales_missing_df['Odometer (KM)'].mean(), inplace=True);
```

```python
# Check how many missing values are there again
car_sales_missing_df.isna().sum()
```

```python
# Delete rows with NaN in the Price column
car_sales_missing_df.dropna(inplace=True)
```

```python
# Check how many missing values are there again
car_sales_missing_df.isna().sum()
```

```python
# Split the data into features and labels
X = car_sales_missing_df.drop("Price", axis = 1)
y = car_sales_missing_df["Price"]
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
rg = RandomForestRegressor(random_state=42)
```

```python
# Future encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
encoder = OneHotEncoder(handle_unknown = "ignore")
# scaler = MinMaxScaler()
scaler = StandardScaler()
categorical_columns = ["Make", "Colour", "Doors"]
numerical_columns = ["Odometer (KM)"]
preprocessor = ColumnTransformer([
    ("Cat",
    encoder,
    categorical_columns),
    ("Num", 
    scaler,
    numerical_columns)
], remainder="passthrough")
transformed_X = preprocessor.fit_transform(X)
pd.DataFrame(transformed_X).head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
rg.fit(X_train, y_train)
```

```python
rg.score(X_test, y_test)
```

```python
for i in range(10, 110, 10):
    rg = RandomForestRegressor(n_estimators=i, random_state=42)
    rg.fit(X_train, y_train)
    print(rg.score(X_test, y_test))
```
