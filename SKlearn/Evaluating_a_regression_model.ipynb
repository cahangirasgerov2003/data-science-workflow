# Evaluating a regression model

```python
import pandas as pd 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
housing_dataset = fetch_california_housing()
housing_dataset
```

```python
housing_df = pd.DataFrame(housing_dataset.data, columns=housing_dataset.feature_names)
housing_df["Target"] = housing_dataset.target
housing_df
```

```python
X = housing_df.drop("Target", axis=1)
y = housing_df["Target"]
y.head(10)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
reg = RandomForestRegressor(n_estimators=100, random_state=42)

reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

# Coefficient of determination

### R^2 = 1 - (sum(sqrt(yi) - sqrt(x1))) / (sum(sqrt(yi) - sqrt(yo)))
* `yi ---- həqiqi dəyərlər`
* `xi ---- praqnozlaşdırılmış dəyərlər`
* `yo ---- həqiqi dəyərlərin ortalaması`

```python
from sklearn.metrics import r2_score
np.mean(y_test)
```

```python
y_test_mean = np.full(len(y_test), np.mean(y_test))
np.shape(y_test), np.shape(y_test_mean), y_test_mean
```

```python
reg.score(X_test, y_test_mean)
```

```python
r2_score(y_test, y_test_mean)
```

# r2-score ---- 1

```python
r2_score(y_test, y_test)
```

# Mean absolute error (MAE)

```python
from sklearn.metrics import mean_absolute_error
y_preds = reg.predict(X_test)
mean_absolute_error(y_test, y_preds)
```

```python
mae = pd.DataFrame(data = {
    "actual data" : y_test,
    "predicted data" : y_preds
})

mae["difference"] = y_test - y_preds
mae
```

```python
mae["difference"].mean()
```

```python
np.abs(mae['difference']).mean()
```

# Mean squared error

```python
mse = mae.copy()
mse["squared_difference"] = np.square(mse["difference"])
mse
```

```python
mse['squared_difference'].mean()
```

```python
mse.iloc[:100, mse.columns.get_loc("squared_difference")] = 20
# mse.iloc[row, column]
# mse.loc[mse.index[:100], "squared_difference"] = 10
mse.head()
```

```python
mse['squared_difference']
```

# Cross validation score and scoring parameter

```python
from sklearn.model_selection import cross_val_score
cv_r2 = cross_val_score(reg, X, y, cv=5, scoring=None)
```

```python
cv_r2
```

```python
print(f"The cross validated coefficient of determination is : {np.mean(cv_r2)}")
```
