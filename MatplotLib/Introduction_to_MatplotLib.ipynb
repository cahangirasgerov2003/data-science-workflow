## State-based API

```python
plt.plot([2, 4, 8, 16], [22, 33, 44, 55])
plt.title("Figure")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.plot([1,2,3], [1,2,3])
plt.title("Figure")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## Statefull API

```python
# First method 
fig = plt.figure();
ax = fig.add_subplot(1,2,2);
# add_subplot ----- 1-ci sətir, 2-ci sütun, 2-ci mövqe 
ax.plot([1, 2, 2], [4, 8, 16])
plt.show();
```

```python
# Second method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
# add_axes ----- left, bottom, width, height ----- 0, 1 arası faiz
ax.plot(x, y);
```

```python
# Recommended way
fig, ax = plt.subplots()
ax.plot(x, y);
```
