# Plotting from Pandas DataFrames

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Plotting from pandas dataframes 5

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease.plot(kind = "scatter", x = "age", y = "chol", figsize = (10, 6), xlim = (20, 90), c = "target");
```

### Object oriented method with mixed pyplot method

```python
fig, ax = plt.subplots(nrows = 1, ncols=1);
heart_disease.plot.scatter(x = "age", y= "chol", c = "target", ax = ax, figsize = (10, 6));
ax.set_facecolor("red")
ax.set_xlim(0, 80);
```

## Plotting from pandas dataframes 6

```python
over_50 = heart_disease[heart_disease['age'] > 50]
len(over_50)
```

### Object oriented method from scratch 

```python
fig, ((ax0), (ax1)) = plt.subplots(nrows = 2, ncols = 1)
scatter = ax0.scatter(x = over_50["age"], y = over_50["chol"], c = over_50["target"])
fig.set_size_inches(10, 20)
ax0.set(title="Heart disease and cholesterol", ylabel="cholesterol")
ax0.legend(*scatter.legend_elements(), title = "Target")
ax0.axhline(y = over_50["chol"].mean(), linestyle = ":", linewidth = 3)
ax0.set_xlim((45, 80))
scatter_thalach = ax1.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50["target"]);
ax1.set(title="Heart disease and maximum heart rate", xlabel = "age", ylabel = "maximum heart rate", xlim=(45, 80))
ax1.legend(*scatter_thalach.legend_elements(), title = "Target")
ax1.axhline(y = over_50["thalach"].mean());
```

## Plotting from pandas dataframes 7

```python
heart_disease.tail()
```

```python
male = heart_disease[heart_disease["sex"] == 0]
male.head()
```

```python
fig, ((ax0, ax1)) = plt.subplots(nrows=2, ncols=1, sharex= True)
fig.set_size_inches(10, 10)
fig.suptitle("Heart disease", horizontalalignment = "center", fontsize = 16, fontweight = "bold", verticalalignment = "center" )
fig.set_edgecolor("blue")
fig.set_linewidth(3)

scatter = ax0.scatter(x = male["age"], y = male["chol"], c = male["target"])
ax0.set(title = "Heart disease and cholesterol", ylabel="Cholesterol")
ax0.axhline(male["chol"].mean(), linestyle = "--", linewidth = 1, color = "red");
ax0.legend(*scatter.legend_elements(), title = "Target");
ax0.grid(visible=True)

scatter = ax1.scatter(x = male["age"], y = male["trestbps"], c = male["target"])
ax1.set(title = "Heart disease and ", xlabel = "Age", ylabel="Resting blood pressure")
ax1.axhline(male["trestbps"].mean(), linestyle = ":", linewidth = 3, color = "green");
ax1.legend(*scatter.legend_elements(), title = "Target");
ax1.grid(visible=True)
```
