```python
# Import relevant libraries
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# State-based way
x = [1, 2, 4]
y = [22, 33, 44]
plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Figure")
plt.show()
```

```python
# Statefull way
## First method
fig = plt.figure(figsize = (2, 2))
ax = fig.add_subplot(1,1,1)
## 1-sətir, 1-sütün olacaq şəkildə 1-in 1-ə bir ərazi yaradır
ax.plot([1, 2, 3], [1, 2, 3])
plt.show()
```

```python
## Second method
fig = plt.figure(figsize = (2, 2))
ax = fig.add_axes((1, 1, 1, 1))
## left, bottom, width, height
ax.plot(x, y)
plt.show()
```

```python
## Recommended mathod
fig, ax = plt.subplots()
fig.set_size_inches(w = 4, h = 2)
ax.plot(x, y)
plt.show()
```

```python
# Figure ----- base canvas
type(fig), type(ax)
```

# MatplotLib example workflow

```python
# 1
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# 2
ndarray_x = np.array([1, 2, 4])
ndarray_y = np.array([22, 44, 88])

# 3
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(w=6, h=6)
ax[0].plot(ndarray_x, ndarray_y)

# 4
ax[0].set_title("Figure")
ax[0].set_xlabel("X-axis")
ax[0].set_ylabel("Y-axis")

# 5 
plt.savefig("./images/Figure.png");
```
