## State-based API ( Plot API )

```python
x = [1, 2, 4, 8]
y = np.random.randint(low=1, high=10, size = (4,) , dtype="int")
x, y
```

```python
plt.plot(x, y)
plt.title("State-based API")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## Statefull API ( Object oriented API )

```python
# 1 method
fig = plt.figure(figsize = (8, 4))
ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)
ax0.plot(x, y)
ax1.plot(x, np.sin(y))
plt.show()
```

```python
# 2 method
fig2 = plt.figure(figsize = (6, 6))
ax2 = fig2.add_axes([0, 0, 0.8, 0.8])
ax2.plot(x, y ** 5)
plt.show()
```

```python
# Recommended method
fig, ax = plt.subplots()
fig.set_size_inches(2, 4)
ax.plot(x, y / x)
ax.set(title="Recommended way", xlabel="X-axis", ylabel="Y-axis")
plt.show()
```

## Anatomy of a matplotlib figure

![](matplotlib-anatomy-of-a-figure.png)

```python
# Line plot
animals_and_avarage_lifespans = {
"Elephant" : 60,
"Dog" : 10,
"Cat" : 12,
"Bear" : 20,
"Giraffe" : 20,
"Dolphin" : 30
}
fig, ax = plt.subplots(nrows=1, ncols=3)
fig.set_size_inches(w=16, h=4)
fig.set_facecolor("#ff0000")
ax[0].set(title="Line plot", xlabel="X-axis", ylabel="Y-axis")
ax[0].plot(np.random.uniform(low=1, high=10, size=(4,)), [1, 8, 16, 32])
ax[1].set(title="Scatter plot", xlabel="X-axis", ylabel="Y-axis")
ax[1].scatter(np.random.rand(100,), np.sin(np.random.rand(100,)))
ax[2].set(title="Hist plot", xlabel="Animals", ylabel="Avarage lifespans")
ax[2].barh(animals_and_avarage_lifespans.keys(), animals_and_avarage_lifespans.values())
plt.show()
```
