## Histogram

```python
import pandas as pd
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
normal_distribution = np.random.randn(1000)
# Normal distribution
fig, ax = plt.subplots()
ax.set(title="Histogram", xlabel="X-axis", ylabel="Y-axis")
ax.hist(normal_distribution)
plt.show()
```

## Subplots

```python
x = np.random.rand(10, 2);
x
```

```python
animals_and_avarage_lifespans = {
    "Elephant": 60,
    "Dog": 10,
    "Cat": 12,
    "Bear": 20,
    "Giraffe": 20,
    "Dolphin": 30,
}
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(12, 6)
ax0.set(title="Line plot", xlabel="X-axis", ylabel="Y-axis")
ax0.plot(np.random.randint(low=10, high=100, size=(10,)), np.random.uniform(low=1, high=10, size=(10,)))
ax1.set(title="Scatter plot", xlabel="X-axis", ylabel="Y-axis")
ax1.scatter(x, np.sin(x))
ax2.set(title="Bar plot", xlabel="X-axis", ylabel="Y-axis")
ax2.barh(animals_and_avarage_lifespans.keys(), animals_and_avarage_lifespans.values())
ax3.set(title="Hist plot", xlabel="X-axis", ylabel="Y-axis")
ax3.hist(np.random.randn(100))
plt.show()
```
