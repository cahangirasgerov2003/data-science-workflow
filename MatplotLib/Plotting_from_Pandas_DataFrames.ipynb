```python
import pandas as pd
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
car_datas = pd.read_csv("car-sales.csv")
car_datas
```

```python
car_datas['Price'] = car_datas["Price"].astype("str").str.replace(r"[^\d.]", "", regex = True).astype("float").astype("int")
```

```python
car_datas
```

```python
car_datas['Odometer (KM)'] = np.floor(car_datas['Odometer (KM)'] / 1.6)
```

```python
car_datas['Odometer (KM)'] = car_datas['Odometer (KM)'].astype("int")
car_datas
```

```python
car_datas = car_datas.rename(columns = {'Odometer (KM)' : 'Odometer (MIL)'})
```

## Plotting from Pandas DataFrames

```python
series = pd.Series(np.random.randn(1000), index=pd.date_range(start="3/5/2026", periods=1000))
series
```

```python
series = series.cumsum()
series
```

```python
ax = series.plot();
ax.set_title("Plotting from Pandas DataFrames");
ax.set_xlabel("X-axis");
ax.set_ylabel("Y-axis");
ax.set_facecolor("yellow");
```

## Avarage price

```python
avarage_price = car_datas.groupby(["Make"])["Price"].mean().reset_index(name = "Avarage price")
```

```python
axes = avarage_price.plot.barh(x="Make", y="Avarage price");
axes.set_xlabel("Avarage price");
```
