# Scatter plot and bar plot

### Line plot

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
ndarray = np.linspace(0, 10, 50)
ndarray[:10]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(2, 2)
ax.plot(ndarray, np.multiply(ndarray, 10))
ax.set(title="Line plot", xlabel="X-axis", ylabel="Y-axis")
plt.show()
```

### Scatter plot

```python
fig, ax = plt.subplots()
ax.set(title="Scatter plot", xlabel="X-axis", ylabel="Y-axis")
ax.scatter(x=ndarray, y=np.sin(ndarray))
plt.show()
```

### Bar plot

```python
animals_and_avarage_lifespans = {
"Elephant" : 60,
"Dog" : 10,
"Cat" : 12,
"Bear" : 20,
"Giraffe" : 20,
"Dolphin" : 30
}
fig, ax = plt.subplots()
fig.set_size_inches(8, 4)
ax.set(title="Bar plot", xlabel="Animals", ylabel="Average lifespans")
ax.bar(animals_and_avarage_lifespans.keys(), animals_and_avarage_lifespans.values())
plt.show()
```
